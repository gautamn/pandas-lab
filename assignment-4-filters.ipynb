```python
#panda do not enforce unique indexes. It is data analyst call
employee = {
    'id':[1, 2, 3],
    'name':['Nitin', 'Gautam', 'Ramesh'],
    'email':['a@x.com', 'b@x.com', 'c@x.com']   
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(employee)
```

```python
#set the filter
filt = (df['name'] == 'Nitin')
```

```python
#apply filter to data frame
df[filt]
```

```python
#loc is used to filter columns by label
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

```python
filt = (df['name'] == 'Nitin') & (df['id'] == 1)
```

```python
df.loc[filt]
```

```python
filt = (df['name'] == 'Nitin') | (df['id'] == 2)
```

```python
df.loc[filt]
```
