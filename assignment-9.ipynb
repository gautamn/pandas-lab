```python
import pandas as pd
import numpy as np
```

```python
employee = {
    'id':[1, 2, 3, 4, 5, 6, 7],
    'name':['Nitin', 'Gautam', 'Ramesh', 'Sapna', np.nan, None, 'NA'],
    'email':['a@x.com', 'b@x.com', 'c@x.com', 'a@x.com', np.nan, None, 'Missing'],
    'age':[35, 65, 40, 36, None, None, 'Missing']
}
```

```python
df = pd.DataFrame(employee)
```

```python
df
```

```python
df.dropna() #np.nan and None removed
```

```python
df.dropna(axis='index', how='any') #axis can be column or index 
#drop rows with any missing values
```

```python
#drop rows with all missing values
df.dropna(axis='index', how='all')
```

```python
#drop columns with all missing values
df.dropna(axis='columns', how='all')
```

```python
#drop columns with any missing values
df.dropna(axis='columns', how='any')
```

```python
df.dropna(axis='column', how='any', subset=['email']) #here how is not doing anything. 
#we are just removing any row with email column value is null
```

```python
df.dropna(axis='index', how='any', subset=['email', 'name']) 
```

```python
df = pd.DataFrame(employee)
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
df.dropna(axis='index', how='any', subset=['email', 'name', 'age']) 
```

```python
df.fillna('MISSING')
```
