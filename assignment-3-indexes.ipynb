```python
#panda do not enforce unique indexes. It is data analyst call
employee = {
    'id':[1, 2, 3],
    'name':['Nitin', 'Gautam', 'Ramesh'],
    'email':['a@x.com', 'b@x.com', 'c@x.com']   
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(employee)
```

```python
df['email']
```

```python
#set email as index.
df.set_index('email')
```

```python
#After above execution Panda does not change the indexes. The default one is still some number
df
```

```python
#To change the indexes permanently we need to set them true.
df.set_index('email', inplace=True)
```

```python
df
```

```python
df.index
```

```python
df.loc['b@x.com']
```

```python
df.loc['b@x.com', 'name']
```

```python
df.iloc[0]
```

```python
df.reset_index(inplace=True)
```

```python
df
```
