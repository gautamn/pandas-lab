```python
df_sur['Country'].value_counts()
```

```python
country_respondents=df_sur['Country'].value_counts()
country_respondents
```

```python
country_uses_python=country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()) #for every country
country_uses_python
```

```python
python_df=pd.concat([country_respondents,country_uses_python ], axis='columns', sort=False )
```

```python
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['PctPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])*100
```

```python
python_df
```

```python
python_df.sort_values(by='PctPython', ascending=False)
```
